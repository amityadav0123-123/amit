**Bar chart**
- in matplotlib plt.bar
- we know that , to draw bar graph we need two things
- lable and its counts
- use value_count()

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10,8))
keys=visa_df['continent'].value_counts().keys()
value=visa_df['continent'].value_counts().values
plt.bar(keys,value,width=0.2) # width is width of bars.
plt.savefig('Continent_Bar_chart.jpg')
plt.show()
```

**pie chart**
- pie chart is available in matplotlib
- it take values and lable

```python
keys=visa_df['continent'].value_counts().keys()
values=visa_df['continent'].value_counts().values
plt.pie(values,labels=keys,
       autopct='%0.3f%%')  # 0.3 means 3 digit after decimal ,0.2 means 2 digit after decimal
plt.show()
```
