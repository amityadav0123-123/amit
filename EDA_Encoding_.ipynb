**Encoding**
- Encoding is a method convert cat columns to num columns
- it is very importand in devolopment of machine learning model
- for this we have some methods
    - map method
    - LableEncoder
    - onehot
    - np.where

**Map**
- here we map cat lables to numerica
- we are following alphbetical order and python index start with zero
- we need to create dictionary lable as key and number as values

```python
visa_df['case_status'].unique()
```

```python
dict1={'Certified':0,'Denied':1}
dict1
```

```python
visa_df['case_status']=visa_df['case_status'].map(dict1)
```

```python
visa_df
```

```python
dict2={'N':0,'Y':1}
visa_df['has_job_experience']=visa_df['has_job_experience'].map(dict2)
```

```python
visa_df
```

```python
for col in cat[1:]:
    uniq=sorted(visa_df[col].unique())
    d={uniq[i]:i for i in range(len(uniq))}
    visa_df[col]=visa_df[col].map(d)
visa_df
```
